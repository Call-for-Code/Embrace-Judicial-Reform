# Statewide Data Analysis

```python
df_data_2 = pd.read_csv('./data/ct_statewide_2020_04_01.csv')
df_data_2.head()
```

```python
sns.countplot(x = df_data_1['outcome'], hue="subject_race", data=df_data_2)
```

```python
sns.countplot(x = 'citation_issued', hue="subject_race", data=df_data_2)
```

```python
percentage_black_citation = df_data_2.loc[df_data_2['subject_race']=='black', 'citation_issued'].sum() / df_data_2['citation_issued'].sum()
```

```python
print("PERCENTAGE OF CITATION FOR BLACK PEOPLE:", percentage_black_citation*100)
```

```python
percentage_white_citation = df_data_2.loc[df_data_2['subject_race']=='white', 'citation_issued'].sum() / df_data_2['citation_issued'].sum()
print("PERCENTAGE OF CITATION FOR WHITE PEOPLE:", percentage_white_citation*100)
```
